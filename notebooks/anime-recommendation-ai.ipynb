```python
import pandas as pd

df = pd.read_csv('../datasets/RecentViewedAnime.csv') 
get_anime = df.head(20)
final_data = get_anime[["ID", "Anime Name", "Rating"]]
print(final_data) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

ypoints = np.array(get_anime[["Rating"]])
xpoints = np.arange(1, len(ypoints) + 1)

plt.plot(xpoints, ypoints)

plt.ylim(0, 10)
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.gca().yaxis.set_major_locator(MultipleLocator(1))

plt.xlabel("Anime ke-")
plt.ylabel("Rating")
plt.title("Rating Anime per Urutan")

plt.show()
```
